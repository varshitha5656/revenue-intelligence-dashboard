```python
import pandas as pd
import matplotlib.pyplot as plt
import os

# load dataset (adjust path if your CSV is elsewhere)
df = pd.read_csv("../data/sales.csv")
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
df.head()
```

```python
monthly = df.resample('M', on='OrderDate')['Sales'].sum()

import matplotlib.pyplot as plt
plt.figure(figsize=(8,4))
monthly.index = monthly.index.strftime('%b\n%Y')
monthly.plot(marker='o')
plt.title('Monthly Revenue Trend')
plt.ylabel('Revenue')
plt.xlabel('Month')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
plt.savefig("../images/monthly_revenue.png", dpi=150, bbox_inches='tight')
```

```python
df['Profit'] = df['Sales'] - df['Cost']
df.head()
```

```python
category_rev = df.groupby('Category')['Sales'].sum()

import matplotlib.pyplot as plt
plt.figure(figsize=(7,4))
category_rev.plot(kind='bar', color='skyblue')
plt.title("Revenue by Category")
plt.ylabel("Revenue")
plt.xlabel("Category")
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
plt.savefig("../images/category_revenue.png", dpi=150, bbox_inches='tight')
```

```python
region_rev = df.groupby('Region')['Sales'].sum()

plt.figure(figsize=(7,4))
region_rev.plot(kind='bar', color='lightgreen')
plt.title("Revenue by Region")
plt.ylabel("Revenue")
plt.xlabel("Region")
plt.grid(axis='y')
plt.tight_layout()
plt.show()
```

```python
plt.savefig("../images/region_revenue.png", dpi=150, bbox_inches='tight')
```

```python
monthly_profit = df.resample('M', on='OrderDate')['Profit'].sum()

plt.figure(figsize=(8,4))
monthly_profit.index = monthly_profit.index.strftime('%b\n%Y')
monthly_profit.plot(marker='o', color='orange')
plt.title("Monthly Profit Trend")
plt.ylabel("Profit")
plt.xlabel("Month")
plt.grid(True)
plt.tight_layout()
plt.show()
```
